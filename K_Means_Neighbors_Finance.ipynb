```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
import pandas as pd
```

```python
datasets = ['BLK']
for dataset in datasets:
    Ticker = yf.Ticker(dataset)
    data = Ticker.history(start='2023-09-01', end='2023-10-13')
    filename=f'{dataset}_data.csv'
    data.to_csv(filename)
    print(f'Download data for {dataset} and saved as {filename}')
```

```python

Ticker='BLK'
start_date = '2023-09-01'
end_date = '2023-10-13'
data = yf.download(Ticker, start=start_date, end=end_date)
```

```python
data.head()
```

```python
data['Up_Down'] = np.where(data['Adj Close'].shift(-1) > data['Adj Close'], 1, -1)
```

```python
data.head()
```

```python
X = data.iloc[: , 1:4].values
y = data.iloc[: , 6].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
